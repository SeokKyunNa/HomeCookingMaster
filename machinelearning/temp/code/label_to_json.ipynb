```python
import xlrd
from collections import OrderedDict
import json

# Windows용 excel_path = '‪C:\\Users\\user\\Desktop\\rpa\\json_to_excel.xlsx'

excel_path = './AI (1).xlsx'
# excel_path = excel_path[1:]
wb = xlrd.open_workbook(excel_path)
sh = wb.sheet_by_index(0)

data_list = []

for rownum in range(1, sh.nrows):
    data = OrderedDict()
    row_values = sh.row_values(rownum)
    data[row_values[0]] = row_values[1]
    
    data_list.append(data)

j = json.dumps(data_list, ensure_ascii=False)

with open('./data_test.json', 'w+') as f:
    f.write(j)
```

```python
import numpy as np
import pandas as pd
import openpyxl 
import json

df = pd.read_excel('AI (1).xlsx', engine='openpyxl')
index_name = {}
for i in range(0, len(df)):
    index_name[int(df[0][i]-1)] = df['food_name'][i]

index_name
json_file = json.dumps(index_name, ensure_ascii=False)
    
json_file
with open('./data_to_json.json', 'w+') as f:
    f.write(json_file)
```
