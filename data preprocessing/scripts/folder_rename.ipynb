```python
import os
```

```python
path = './img'
```

```python
for i in range(2, 1856):
    file_list = os.listdir(f'{path}/{i}')
    for file in file_list:
        new_file = file.replace(f'{i - 1}_', f'{i}_')
        os.rename(f'{path}/{i}/{file}', f'{path}/{i}/{new_file}')
    print(f'folder {i} 완료')
```
