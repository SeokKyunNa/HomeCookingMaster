```python
import os
import re
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sample_text = recipe_df.iloc[15]['재료']
sample_text
```

```python
m = re.compile(r"\<.*?\>")

new_content = re.sub(r"\[.*?\]","|",sample_text)
new_content = re.sub(r"\(.*?\)","",new_content)
new_content
```

```python
text_list = new_content.split('|')
text_list_2 = []
for text in text_list:
    if text != '':
        text_list_2.append(text.strip())
text_list_2
```

```python
for text in text_list_2:
    temp_text = text.split(' ')
    amount = temp_text[-1]
    ingredient = " ".join(temp_text[:-1])
    print(ingredient, amount)
```

```python
def ingredient_data_preprocessing (index):
    ingredient_origin = recipe_df.iloc[index]['재료']
    recipe_id = recipe_df.iloc[index]['recipe_id']

    #대괄호 처리
    ingredient_pre_1= re.sub(r"\[.*?\]","|",ingredient_origin)
    #소괄호 처리
    ingredient_pre_2 = re.sub(r"\(.*?\)","",ingredient_pre_1)

    # | 기준으로 나누기
    ingredient_pre_3 = ingredient_pre_2.split('|')
    ingredient_pre_4 = []
    for ingredient in ingredient_pre_3:
        if ingredient != '':
            ingredient_pre_4.append(ingredient.strip())

    # dataframe 만들기
    temp_ingredient_df = pd.DataFrame(columns=['recipe_id', '재료명', '재료양'])

    # 재료명이랑 재료양 나눠서 dataframe에 넣기
    for ingredient in ingredient_pre_4:
        temp_ingredient = ingredient.split(' ')
        ingredient_dic = {'recipe_id' : recipe_id, "재료명" : " ".join(temp_ingredient[:-1]), "재료양" : temp_ingredient[-1]}
        print(ingredient_dic)
        temp_ingredient_df.append(ingredient_dic, ignore_index=True)

    print(f'레시피 {recipe_id} 재료 추가')
    print(temp_ingredient_df)
    return temp_ingredient_df
```

```python
ingredient_result_df = pd.DataFrame(columns=['recipe_id', '재료명', '재료양'])
```

```python
for i in range(recipe_df.shape[0]):
    print({recipe_df['recipe_id'][i]}, " 레시피 재료 찾는 중...")
    ingredient_result_df = pd.concat([ingredient_result_df, ingredient_data_preprocessing(i)], ignore_index=True)
ingredient_result_df.head()
```
