```python
def find_recipe_data(index):
    if df_label_origin['소분류구분'][index] in convert_label_dic.keys():
        df_temp = df_recipe[df_recipe['요리명'].str.contains(convert_label_dic[df_label_origin['소분류구분'][index]])]
    else:
        df_temp = df_recipe[df_recipe['요리명'].str.contains(df_label_origin['소분류구분'][index])]
    
    if df_label_origin['소분류구분'][index] in del_label_dic.keys():
        for word in del_label_dic[df_label_origin['소분류구분'][index]]:
            df_temp = df_temp[~df_temp['요리명'].str.contains(word)]
    df_temp.insert(0,'요리',df_label_origin['소분류구분'][index])
    df_temp.insert(0,'food_id',df_label_origin.index[index])
    return df_temp.iloc[0:5]
```

```python
df_recipe_result = pd.DataFrame(columns=['food_id', '요리',	'레시피 고유 번호',	'레시피 제목', '요리명', '요리 소개','스크랩수', '재료', '요리인분', '요리 난이도', '요리 시간'])
df_recipe_result
```

```python
for i in range(df_label_origin.shape[0]):
    print({df_label_origin['소분류구분'][i]}, " 레시피 찾는중...")
    df_recipe_result = pd.concat([df_recipe_result, find_recipe_data(i)], ignore_index=True)
df_recipe_result.head()
```
